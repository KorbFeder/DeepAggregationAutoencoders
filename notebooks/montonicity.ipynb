```python
x = torch.Tensor([0.1, 0.2, 0.2, 0.8, 0.1, 0.7])
minimums = []
indices = []
for _ in range(len(x)):
	value, index = torch.min(x, dim=-1)
	minimums.append(value)
	indices.append(index)
	x = torch.cat([x[:index], x[index+1:]])

plt.plot(minimums)
plt.show()
```

```python
y = torch.Tensor([0.3])
mse = torch.nn.MSELoss()
losses = []
for m in minimums:
	losses.append(mse(y, torch.Tensor(m)))
	
plt.cla()
plt.plot(losses)
plt.show()
```
